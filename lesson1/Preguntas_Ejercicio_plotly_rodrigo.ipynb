#### Realiza una gráfica que muestre la proporción de viajes totales de cada organismo

```python
viajes = data[['organismo','afluencia_total_preliminar']].groupby(["organismo"]).sum().reset_index().\
                sort_values( by="afluencia_total_preliminar",ascending=False)
viajes['afluencia_total_preliminar']= np.round((viajes['afluencia_total_preliminar']*100)/\
                                               sum(viajes['afluencia_total_preliminar']),2)
viajes
```

```python
# fig = px.funnel(viajes, x='organismo',y = 'afluencia_total_preliminar')
# fig = px.funnel(viajes, x='afluencia_total_preliminar',y = 'organismo' ,orientation='h')
fig = px.pie(viajes, values='afluencia_total_preliminar', names='organismo',\
             title='Proporción de viajes por organismo')
fig.show()
```

```python
# Varias gráficas
x = viajes['afluencia_total_preliminar']
y = viajes['organismo']

fig = make_subplots(rows=1, cols=2, specs=[[{'type':'pie'},{'type':'funnel'}]],horizontal_spacing=0.15)

fig.add_trace(go.Pie(values=x,labels=y,domain=dict(x=[0,0.5]),hole=0.3),1,1)
fig.add_trace(go.Funnel(x=x,y=y,name='Proporcion de viajes total',\
                        textinfo = "value+percent total"),\
              1,2)

fig.show()
```
